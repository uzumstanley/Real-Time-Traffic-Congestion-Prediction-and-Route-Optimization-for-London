```python
import pandas as pd
import dash
from dash import dcc, html
import plotly.express as px
from dash.dependencies import Input, Output

# Load and process the dataset
#data_path = 'Strategic_Monitoring_Areas.csv'  # Replace with your actual file path
#data = pd.read_csv(data_path)

# Deduplication
data = data.drop_duplicates()

# Summary statistics grouped by ZONE and VALUE
summary_stats = data.groupby(['ZONE', 'VALUE']).size().reset_index(name='Count')

# Initialize Dash app
app = dash.Dash(__name__)
app.title = "Strategic Monitoring Areas Dashboard"

# Layout of the app
app.layout = html.Div([
    html.H1("Strategic Monitoring Areas Dashboard", style={'textAlign': 'center'}),

    # Summary statistics table
    html.H2("Summary Statistics"),
    html.Div(id='summary-table', children=[
        html.Table([
            html.Thead(html.Tr([html.Th(col) for col in summary_stats.columns])),
            html.Tbody([
                html.Tr([html.Td(row[col]) for col in summary_stats.columns]) for _, row in summary_stats.iterrows()
            ])
        ])
    ]),

    # Visualization controls
    html.H2("Visualizations"),
    html.Div([
        html.Label("Select Visualization Type:"),
        dcc.Dropdown(
            id='chart-type',
            options=[
                {'label': 'Pie Chart (Area Distribution)', 'value': 'pie'},
                {'label': 'Bar Chart (Zone Counts)', 'value': 'bar'}
            ],
            value='pie',
            clearable=False
        )
    ]),

    # Visualization output
    dcc.Graph(id='chart-output')
])

# Callback to update the chart
@app.callback(
    Output('chart-output', 'figure'),
    [Input('chart-type', 'value')]
)
def update_chart(chart_type):
    if chart_type == 'pie':
        fig = px.pie(data, names='ZONE', title='Area Distribution by Zone')
    elif chart_type == 'bar':
        zone_counts = data['ZONE'].value_counts().reset_index()
        zone_counts.columns = ['ZONE', 'Count']
        fig = px.bar(zone_counts, x='ZONE', y='Count', title='Zone Counts')
    else:
        fig = {}
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
